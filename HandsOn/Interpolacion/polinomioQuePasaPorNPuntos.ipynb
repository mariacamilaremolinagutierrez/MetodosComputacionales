```python
def polinomio(x,n,x_in,y_in):
    suma=0
    for i in range(n):
        x_i=x_in[i]
        r=1
        for j in range(n):
            x_j=x_in[j]
            if(x_i!=x_j):
                r=r*(x-x_j)/(x_i-x_j)
        suma+=r*y_in[i]
    
    return suma
```

```python
#Ejemplo con la funcion a continuacion

def p(x):
    return 1.0/(1.0+25.0*x**2)
```

```python
a = linspace(-1.0,1.0,3)
b = linspace(-1.0,1.0,5)
c = linspace(-1.0,1.0,9)
```

```python
scatter(a,p(a), color="black", alpha=1.0, s=30)
scatter(b,p(b), color="red", alpha=0.7,s=120)
scatter(c,p(c), color="green", alpha=0.4,s=500)
```

```python
puntos=linspace(-1,1,500)
poli3p=polinomio(puntos,3,a,p(a))
poli5p=polinomio(puntos,5,b,p(b))
poli9p=polinomio(puntos,9,c,p(c))
```

```python
scatter(puntos,poli3p, color='blue')
scatter(a,p(a), color='magenta')
plot(puntos,p(puntos), color='green')
```
