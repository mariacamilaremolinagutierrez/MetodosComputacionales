```python
%pylab inline
import scikits.audiolab as audio
from scipy.fftpack import fft, fftfreq
```

```python
#Se lee el archivo completo
input_signal, sampling_rate, enc = audio.wavread("audio.wav")
time_array = arange(0, len(input_signal)/float(sampling_rate), 1/float(sampling_rate))

plot(time_array,input_signal)
```

```python
t=len(time_array)

tiempo=time_array[t/7:4.6*t/6]
amplitud=input_signal[t/7:4.6*t/6]

plot(tiempo,amplitud)
xlabel("time(s)", fontsize=20)
ylabel("Amplitude", fontsize=20)
```

```python
#Numero de puntos
n = len(tiempo)
print n

#Tranformada de Fourier normalizada
fft_x = fft(amplitud) / n

#dt, cada cuando hay un valor:
dt = tiempo[1]-tiempo[0]
print dt

#Se recuperan las frecuencias
freq = fftfreq(n, dt) 
```

```python
plot(freq,abs(fft_x))
xlim([-0.05,5000])
```

```python
#Esta es claramente la frecuencia que quiero quitar, por lo tanto la dejo en 0 y reconstruyo.
plot(freq,abs(fft_x))
xlim([990,1020])
```

```python
indices_pito_1=where(freq>=995)
indices_pito_2=where(freq<=1005)
indices_pito_3=where(freq<=-995)
indices_pito_4=where(freq>=-1005)

indices_pito=intersect1d(indices_pito_1[0],indices_pito_2[0])
indices_pito_neg=intersect1d(indices_pito_3[0],indices_pito_4[0])

fft_x[indices_pito]=0
fft_x[indices_pito_neg]=0

plot(freq,abs(fft_x))
xlim([-0.05,1050])
```

```python
#Ahora se reconstruye

filtrada= ifft(fft_x) 
plot(tiempo,real(filtrada))
```

```python
audio_filtrado = audio.Sndfile("audio_filtrado.wav", 'w', audio.Format('wav'), 1, sampling_rate)
audio_filtrado.write_frames(real(filtrada))
audio_filtrado.close()
```
