```python
%pylab inline
import scikits.audiolab as audio
from scipy.fftpack import fft, fftfreq
```

```python
#Se lee el archivo completo
input_signal, sampling_rate, enc = audio.wavread("mc_normal.wav")
time_array = arange(0, len(input_signal)/float(sampling_rate), 1/float(sampling_rate))

plot(time_array,input_signal)
xlabel("time(s)", fontsize=20)
ylabel("Amplitude", fontsize=20)
```

```python
t=len(time_array)

tiempo=time_array[1.7*t/4:3.4*t/4]
amplitud=input_signal[1.7*t/4:3.4*t/4]

plot(tiempo,amplitud)
xlabel("time(s)", fontsize=20)
ylabel("Amplitude", fontsize=20)
```

```python
#Numero de puntos
n = len(tiempo)
print n

#Tranformada de Fourier normalizada
fft_x = fft(amplitud) / n

#dt, cada cuando hay un valor:
dt = tiempo[1]-tiempo[0]
print dt

#Se recuperan las frecuencias
freq = fftfreq(n, dt) 
```

```python
plot(freq,abs(fft_x))
xlim([-0.05,12000])
```

```python
#Disminuyo todas las frecuencias

freq=freq/10
plot(freq,abs(fft_x))
xlim([-0.05,1200])
```

```python
#Ahora se reconstruye

agravada= ifft(fft_x) 
plot(tiempo,real(agravada))
xlabel("time(s)", fontsize=20)
ylabel("Amplitude", fontsize=20)
```

```python
audio_filtrado = audio.Sndfile("mc_darthvader.wav", 'w', audio.Format('wav'), 1, sampling_rate)
audio_filtrado.write_frames(real(agravada))
audio_filtrado.close()
```
