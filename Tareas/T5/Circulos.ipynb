## $R[j]=\sum{x_i*x_{i-j}}$

```python
#Funciones

def autocorrelacion_x(j,n,arreglo_x):
    R=0
    for i in range(n):
        lim=i-j
        if(lim>=0):
            R+=arreglo_x[i]*arreglo_x[lim]
            
    return R
```

```python
lim=num_puntos
nums_lim=range(0,lim)
Rs=zeros(lim)

for j in range(lim):
    Rs[j]=autocorrelacion_x(j, num_puntos, xs)
```

```python
plot(nums_lim,Rs)
```

```python
plot(nums_lim,Rs)
xlim([2500,3000])
```
