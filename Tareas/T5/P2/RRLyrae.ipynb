```python
from scipy.fftpack import fft, fftfreq
```

```python
data = loadtxt("RR_Lyrae_template.dat")

n= size(data[:,0])

dt = 4.0/size(data[:,1])

chisquared = empty((12))

plot(data[:,0], data[:,1])
xlabel("Phase", fontsize=20)
ylabel("Amplitude", fontsize=20)
```

```python
fft_x = fft(data[:,1]) # FFT Normalized
freq = fftfreq(n, dt) # Recuperamos las frecuencias

plt.plot(freq,np.abs(fft_x))
xlabel("Frequency", fontsize=20)
ylabel("|fft_x|", fontsize=20)
axis((-12.5,12.5,0,2000))

print freq
```

### Componentes de Fourier para $(N=0)$

```python
fft_x = fft(data[:,1]) # FFT Normalized
freq = fftfreq(n, dt) # Recuperamos las frecuencias
#plt.plot(freq,np.abs(fft_x))

#set to 0 the values of the transform that are high:
freq_cut = 0
fft_x[abs(freq) > freq_cut] = 0

clean_f = ifft(fft_x)
plot(data[:,0],clean_f,c='r')
plot(data[:,0],data[:,1])
xlabel("Phase", fontsize=20)
ylabel("Amplitude", fontsize=20)
```

### $\chi^2$ para $N=0$

```python
chisquared[0]=abs(sum((data[:,1]-clean_f)**2))
print chisquared[0]
```
