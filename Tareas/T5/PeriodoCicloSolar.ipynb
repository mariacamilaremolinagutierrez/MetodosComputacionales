```python
%pylab inline
from scipy.interpolate import interp1d
from scipy.fftpack import fft, fftfreq
```

```python
matriz = loadtxt('./sparse_sample_monthrg.dat')
anios=matriz[:,0]
meses=matriz[:,1]
dias=matriz[:,2]
manchas=matriz[:,3]
```

```python
num_anios=len(anios)

nuevos_anios=[]
nuevas_manchas=[]

num_manchas=0

#Quito datos con -99 y defino los años como año+mes/12
for i in range(num_anios):
    if(dias[i]!=0): 
        a=anios[i]+meses[i]/12
        m=manchas[i]
        num_manchas+=1
        nuevos_anios.append(a)     
        nuevas_manchas.append(m) 
    
nuevos_anios=array(nuevos_anios)    
nuevas_manchas=array(nuevas_manchas)
```

```python
#Defino la interpolacion lineal
interpolacion = interp1d(nuevos_anios, nuevas_manchas)
```

```python
#Interpolo los datos y grafico

anios_interpolados = linspace(1613,1995,(1995-1613)*12)
manchas_interpoladas = interpolacion(anios_interpolados)

scatter(nuevos_anios, nuevas_manchas)
plot(anios_interpolados, manchas_interpoladas)

print anios_interpolados, manchas_interpoladas
```

```python
num_manchas = len(manchas_interpoladas)

#Tranformada de Fourier normalizada
fft_x = fft(manchas_interpoladas) / num_manchas 

#dt, cada cuando hay un valor:
dt=anios_interpolados[2]-anios_interpolados[1]

#Se recuperan las frecuencias
freq = fftfreq(num_manchas, dt) 

plot(freq,abs(fft_x))
xlim([-0.05,0.5])
```

```python
#Se obtiene la frecuencia mas alta que no sea 0. 
i_f0=argmax(abs(fft_x))
fft_x[i_f0]=0

i_max=argmax(abs(fft_x))
f_max=abs(freq[i_max])

#While hasta que la amplitud de la frecuencia maxima deje de ser parecida a la amplitud en f=0
while(f_max<0.05):
    i_f_quitar=argmax(abs(fft_x))
    fft_x[i_f_quitar]=0
    i_max=argmax(abs(fft_x))
    f_max=abs(freq[i_max])

#El periodo es igual a 1/frecuencia
T=1/f_max

print "El ciclo del Sol es:", T, "años"
```

### Se puede ver claramente como la ausencia de datos entre 1600-1730 y el hueco entre 1800 y 1850 afectan el ciclo, y por ende la transformada de fourier, por tal razon, se calculara el ciclo con los datos desde 1850 hasta el final. 

```python
indices_nuevos=where(anios_interpolados>=1850)

inicio=indices_nuevos[0][0]

nuevos_anios_interpolados=anios_interpolados[inicio:]
nuevas_manchas_interpoladas=manchas_interpoladas[inicio:]

plot(nuevos_anios_interpolados, nuevas_manchas_interpoladas)
```

```python
num_manchas=len(nuevas_manchas_interpoladas)

#Tranformada de Fourier normalizada
fft_x = fft(nuevas_manchas_interpoladas) / num_manchas

#dt, cada cuando hay un valor:
dt=nuevos_anios_interpolados[2]-nuevos_anios_interpolados[1]

#Se recuperan las frecuencias
freq = fftfreq(num_manchas, dt) 

plot(freq,abs(fft_x))
xlim([-0.05,0.6])
```

```python
#Se obtiene la frecuencia mas alta que no sea 0
i_f0=argmax(abs(fft_x))
fft_x[i_f0]=0

i_max=argmax(abs(fft_x))
f_max=abs(freq[i_max])

#While hasta que la amplitud de la frecuencia maxima deje de ser parecida a la amplitud en f=0
while(f_max<0.05):
    i_f_quitar=argmax(abs(fft_x))
    fft_x[i_f_quitar]=0
    i_max=argmax(abs(fft_x))
    f_max=abs(freq[i_max])

#El periodo es igual a 1/frecuencia
T_nuevo=1/f_max

print "El ciclo del Sol es:", T_nuevo, "años"
```
