```python
data = np.loadtxt("movimiento.dat")
data = data.T
plot(data[0],data[1])
```

```python
# Calcula el likelihood de una grupo de valores calculados respecto a los valores reales
def the_hood( real_y, model_y ):
    lol = 0.5*np.sum((real_y-model_y)**2)
    return lol
```

```python
# Da los valores dado el modelo y un punto o grupo de puntos
def the_model_y( x, A, B, C ):
    y = A*x*x + B*x + C
    return y
```

```python

# Cadena de Markov para encontrar el mejor modelo (con mas likelihood)
# nami =  numero de iteraciones
# choper = desviacion de la gausiana de los pasos aleatorios
def MCMC( nami , choper, data ):
    
    # Inicializa las contenedoras de las variables del fit
    A_model = []
    B_model = []
    C_model = []
    luffylihood = []
    
    # Los primeros parametros del modelo --> aleatorio
    A_model.append(np.random.random())
    B_model.append(np.random.random())
    C_model.append(np.random.random())
    
    # Calcula N del primer modelo aproximado
    funcN_init = the_model_y(data[0], A_model[0], B_model[0],C_model[0])
    # El primer likelihood
    luffylihood.append(the_hood(data[1], funcN_init))
    i = 0
    for i in range(nami):
        
        # Prueba variando los parametros
        A_temp = np.random.normal(A_model[i], choper)
        B_temp = np.random.normal(B_model[i], choper)
        C_temp = np.random.normal(C_model[i], choper)
        
        # Calcula la funcion con los nuevos parametros
        newFunc = the_model_y(data[0], A_temp, B_temp, C_temp)
        
        # Calcula la nueva likelihood
        luffy = the_hood( data[1], newFunc )
        
        # Si la likelihood es mejor que la anterior, actualiza los parametros
        if luffy < luffylihood[i] :
            A_model.append(A_temp)
            B_model.append(B_temp)
            C_model.append(C_temp)
            luffylihood.append(luffy)
          
        
        # Si no, se deja a la suerte
        else:    
            robin = np.random.random()
            sanji = np.exp(-(luffy - luffylihood[i]))    
            if robin <= sanji:
                A_model.append(A_temp)
                B_model.append(B_temp)
                C_model.append(C_temp)
                luffylihood.append(luffy)
            else:
                A_model.append(A_model[i])
                B_model.append(B_model[i])
                C_model.append(C_model[i])
                luffylihood.append(luffylihood[i])

    return np.array([luffylihood, A_model, B_model, C_model])
```

```python
zoro = MCMC(100000, 0.01, data)
```

```python
plot(np.linspace(0,1,len(zoro[0])), zoro[0])
```
