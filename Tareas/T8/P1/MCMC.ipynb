```python
def myFunc( x ):
    return np.exp(- ((x-30.0)**2)/100 )
```

```python
xsp = np.linspace(0,60,1000)
plot(xsp,myFunc(xsp))
```

```python
# Retorna una lista histograma aproximando la función MyFunc
# Sigma: el tamaño de los pasos
# n_points: el numero de puntos en el histograma
# xsp: un array con el rango de la función en donde se quiere aproximar
def MCMC_1D( n_points, sigma , xsp ):
    # La lista donde se guardara el histograma
    ListLOLOL = []
    # El punto inicial
    x0 = np.min(xsp) + (np.max(xsp)-np.min(xsp))*np.random.random()
    F0 = myFunc(x0)
    ListLOLOL.append(x0)
    
    for i in range(n_points):
        xtemp = np.random.normal(x0,sigma)
        Ftemp = myFunc(xtemp)

        if Ftemp >= F0 :
            F0 = Ftemp
            x0 = xtemp
            ListLOLOL.append(x0)

        else:

            alpha = np.random.random()
            #print alpha, Ftemp/F0
            if alpha <= Ftemp/F0:
                F0 = Ftemp
                x0 = xtemp
                ListLOLOL.append(x0)

            else: 
                ListLOLOL.append(x0)
    
    return np.array(ListLOLOL)
    
```

```python
# Plotea
f = myFunc(xsp)
norm = sum(f*(xsp[1]-xsp[0]))
plot(xsp,f/norm, linewidth=1, color='r')

x_walk = MCMC_1D(100000, 1, xsp )
#print x_walk
count, bins, ignored = plt.hist(x_walk, 1000, normed=True)

#fig = figure(1, figsize=(9.5,6.5))
plt.xlabel('x')
plt.ylabel('p(x)')
ax = axes()
ax.set_xlim([-0,60])
ax.set_ylim([0.0,0.1])
```
