```python
#Datos

num_anios=24
num_filas=0
num_columnas=0

with open('homicidios.1990.a.2013.csv', 'rb') as archivo:
    lector = csv.reader(archivo)
    i=0
    for fila in lector:
        if(i==0):
            for columna in fila:
                num_columnas+=1
        num_filas+=1
        i+=1
        
#El archivo es: CODIGO - MUNICIPIO - DEPARTAMENTO - HOMICIDIOSAÑO - POBLACIONAÑO - TASAAÑO

#Voy a meter los datos en  matrices: matrizC(cod), matrizM(mun), matrizD(dep) y matriz(HOM-POB-TAS)

matrizC=[]
matrizM=[]
matrizD=[]
matriz=ones((num_filas-1,num_columnas-3))       

with open('homicidios.1990.a.2013.csv', 'rb') as archivo:
    lector = csv.reader(archivo)
    i=0
    for fila in lector:
        if(i!=0):         
            j=0
            for columna in fila:
                if(j==0):
                    #meto en matrizC
                    matrizC.append(int(columna))
                elif(j==1):
                    #meto en matrizM
                    matrizM.append(columna)
                elif(j==2):
                    #meto en matrizD
                    matrizD.append(columna)
                else:
                    #meto en la matriz
                    indice=(i-1,j-3)
                    if (columna=='NA' or columna=='#DIV/0!'):
                        matriz[indice]=-1
                    else:
                        matriz[indice]=columna
                j+=1
        i+=1
        
#Grafica de codigos de municipios contra tasa de homicidios (eje x: codigo / eje y: tasa)
for i in range(num_anios):
    scatter(matrizC,matriz[:,3*i+2],color=random.rand(6,1))
```

```python
#Ahora defino la funcion sigmaij que me da el elemento en la posicion ij de la matriz de covarianza

def sigmaij(d_subi,d_subj,M):
    s=0
    k=0
    for k in range(M):
        dsik=d_subi[k]
        dsim=mean(d_subi)
        dsjk=d_subj[k]
        dsjm=mean(d_subj)
        if(dsik==-1):
            M-=1
        elif(dsjk==-1):
            M-=1
        else:
            s+=(dsik-dsim)*(dsjk-dsjm)/(M-1)
    return s
```

```python
#Matriz de Covarianza

#Dimension= num_anios
#M=num_filas-1

matriz_covarianza=zeros((num_anios,num_anios))

#A llenar la matriz de covarianza
i=0
for i in range(num_anios):
    j=0
    for j in range(num_anios):
        indice=(i,j)
        matriz_covarianza[indice]=sigmaij(matriz[:,3*i+2],matriz[:,3*j+2],num_filas-1)
        
imshow(log10(abs(matriz_covarianza))+1E-30)
```

```python
#Creo los valores y vectores propios
valores_propios, vectores_propios=linalg.eig(matriz_covarianza)

#Hallo la suma de los valores propios y el porcentaje que aporta cada uno

sumaVP=0
i=0
for i in range(num_anios):
    sumaVP+=valores_propios[i]
    
valorActual=0
i=0
for i in range(num_anios):
    valorActual+=valores_propios[i]
    print "Hasta el valor en la posición",i, "se ha aportado",  valorActual*100/sumaVP, "% al valor de la traza"
```

## b) K-Clusters

```python
#Funciones

#Recibe una matriz de centros(R_n x num_centros), y una matriz de puntos(R_n x num_puntos) para hallarle los nuevos centros. 
def nuevosCentros(centros,num_centros, puntos, num_puntos, R_n):
    
    #Las distancias de cada uno de los puntos al centro actual (en orden de lectura de un parrafo).
    #Linea n: distancias de cada punto al centro n
    #Columna j: distancias del punto j a cada centro
    distancias=zeros((num_puntos,num_centros))
    
    i=0
    for i in range(num_centros):
        
        #El primer centro de la lista que me entra por parametro
        centroActual=centros[:,i]
        
        j=0
        for j in range(num_puntos):
            
            puntoActual=puntos[:,j]
            
            d_2=0
            k=0
            for k in range(R_n):
                centro_coord_n=centroActual[k]
                punto_coord_n=puntoActual[k]          
                d_2+=(centro_coord_n-punto_coord_n)**2
                
            d=sqrt(d_2)
            
            indice=(j,i)
            distancias[indice]=d
            
    #Ahora hallo los promedios y nuevos centros
    
    #Primero lleno en nuevos centros las sumas de las coordenadas de los puntos que le pertenecen a el    
    #Para eso, en mi matriz de distancias miro fila por fila, cojo el menor valor (menor distancia) y 
    #ese indice ij, va a ser la distancia del punto i al centro j, es decir el punto asociado a ese centro
    
    paraDividir=zeros(num_centros)
    nuevosCentros=zeros((R_n,num_centros))
    
    i=0   
    for i in range(num_puntos):
        fil=distancias[i,:]
        ind=argmin(fil)
        
        j=0
        for j in range(R_n):        
            nuevosCentros[(j,ind)]+=puntos[(j,i)]
            
        paraDividir[ind]+=1
    
    #Ahora los promedio diviendo en el numero de puntos que sumo para ese centro
    i=0
    for i in range(num_centros):
        div=paraDividir[i]
        
        if(div!=0):
            
            j=0
            for j in range(R_n):
                nuevosCentros[(j,i)]=nuevosCentros[(j,i)]/div
        
        else:
            
            j=0
            for j in range(R_n):
                nuevosCentros[(j,i)]=centros[(j,i)]
   
    #Ya tengo mis nuevos centros :D  
        
    return nuevosCentros      
```

```python
#Voy a definir la funcion que itere nuevosCentros hasta que sean iguales iteraciones n y n+1 

def nuevosCentrosIterados(centros,num_centros, puntos, num_puntos, R_n):
    
    nc_anterior=nuevosCentros(centros,num_centros, puntos, num_puntos, R_n)
    nc_actual=nuevosCentros(nc_anterior,num_centros, puntos, num_puntos, R_n)
    
    ant_igual_actual=False
    
    while(ant_igual_actual==False):
        nc_anterior=nc_actual
        nc_actual=nuevosCentros(nc_anterior,num_centros, puntos, num_puntos, R_n)
        
        contador=0
        i=0
        for i in range(R_n):
            j=0
            for j in range(num_centros):
                indice=(i,j)             
                if(nc_anterior[indice]==nc_actual[indice]):
                    contador+=1
                    
        if(contador==R_n*num_centros):
            ant_igual_actual=True
            break
        
    return nc_actual

#Función para calcular K centros aleatorios
def centrosAletorios(R_n,K):
    cA=ones((R_n,K))
    
    i=0
    for i in range(K):
        j=0
        for j in range(R_n):
            indice=(j,i)
            cA[indice]=random.random()*(10**(3*random.random()))
            
    return cA
```

### Ahora a hallar los K-Clusters para K=2,3,4,5
