# a) PCA

```python
#Funciones

#sigmaij: me da el elemento en la posicion ij de la matriz de covarianza
def sigmaij(d_subi,d_subj,M):
    s=0
    k=0
    for k in range(M):
        dsik=d_subi[k]
        dsim=mean(d_subi)
        dsjk=d_subj[k]
        dsjm=mean(d_subj)
        if(dsik==-1):
            M-=1
        elif(dsjk==-1):
            M-=1
        else:
            s+=(dsik-dsim)*(dsjk-dsjm)/(M-1)
    return s
```

```python
#Matriz de Covarianza

#Dimension= num_columnas
#M=num_municipios

matriz_covarianza=zeros((num_columnas,num_columnas))

#A llenar la matriz de covarianza
i=0
for i in range(num_columnas):
    j=0
    for j in range(num_columnas):
        indice=(i,j)
        matriz_covarianza[indice]=sigmaij(matriz_tasas[i,:],matriz_tasas[j,:],num_municipios)
```

```python
#Grafiquemos la matriz de covarianza

imshow(log10(abs(matriz_covarianza))+1E-30)

print matriz_covarianza
```

```python
#Creo los valores y vectores propios
valores_propios, vectores_propios=linalg.eig(matriz_covarianza)

print "Los vectores propios son: \n",vectores_propios
print "\n"

#Hallo la suma de los valores propios y el porcentaje que aporta cada uno

sumaVP=0
i=0
for i in range(num_columnas):
    sumaVP+=valores_propios[i]
    
valorActual=0
i=0
for i in range(num_columnas):
    valorActual+=valores_propios[i]
    print "Hasta el valor en la posición",i, "se ha aportado",  valorActual*100/sumaVP, "% al valor de la traza"
```

### RESPUESTA: Se puede ver que en la posicion 1 ya se puede tener el 89% del valor de la traza, esto quiere decir que con los 2 primeros autovectores se puede recrear nuestros datos con una muy buena aproximacion. Entonces si es posible disminuir la dimensionalidad de los datos de R_4 a R_2

```python
#Ahora voy a disminuir la dimensionalidad de los vectores propios a 2x2 y a crear misVectores :)

vectoresNecesarios=2
misVectores=zeros((vectoresNecesarios,vectoresNecesarios))

i=0
for i in range(vectoresNecesarios):
    j=0
    for j in range(vectoresNecesarios):
        indice=(i,j)
        misVectores[indice]=vectores_propios[indice]       
        
print "Los nuevos vectores base de mi informacion son: \n", misVectores
        
misVectores=matrix(misVectores)
```

```python
#Ahora hayo una matriz de misCoeficientes, de 2x4 que tiene en cada columna los coeficientes de cada parametro 

misCoeficientes=zeros((vectoresNecesarios,num_columnas))

i=0
for i in range(num_columnas):
    j=0
    col_i_Cov=zeros(vectoresNecesarios)
    
    for j in range(vectoresNecesarios):
        indice=(i,j)
        col_i_Cov[j]=matriz_covarianza[indice]
        
    col_i_Cov=matrix(col_i_Cov)
    coeficientesColumnai=misVectores*col_i_Cov.T   
    
    j=0
    for j in range(vectoresNecesarios):
        indice=(j,i)
        misCoeficientes[indice]=coeficientesColumnai[j]
        
print "Los coeficientes necesarios para expresar mis datos como combinacion lineal de misVectores son: \n",misCoeficientes
```

# b) K-Clusters

```python
#Funciones

#nuevosCentros: Recibe una matriz de centros(R_n x num_centros), y una matriz de puntos(R_n x num_puntos) 
#para hallarle los nuevos centros. 
def nuevosCentros(centros,num_centros, puntos, num_puntos, R_n):
    
    #Las distancias de cada uno de los puntos al centro actual (en orden de lectura de un parrafo).
    #Linea n: distancias de cada punto al centro n
    #Columna j: distancias del punto j a cada centro
    distancias=zeros((num_puntos,num_centros))
    
    i=0
    for i in range(num_centros):
        
        #El primer centro de la lista que me entra por parametro
        centroActual=centros[:,i]
        
        j=0
        for j in range(num_puntos):
            
            puntoActual=puntos[:,j]
            
            d_2=0
            k=0
            for k in range(R_n):
                centro_coord_n=centroActual[k]
                punto_coord_n=puntoActual[k]          
                d_2+=(centro_coord_n-punto_coord_n)**2
                
            d=sqrt(d_2)
            
            indice=(j,i)
            distancias[indice]=d
            
    #Ahora hallo los promedios y nuevos centros
    
    #Primero lleno en nuevos centros las sumas de las coordenadas de los puntos que le pertenecen a el    
    #Para eso, en mi matriz de distancias miro fila por fila, cojo el menor valor (menor distancia) y 
    #ese indice ij, va a ser la distancia del punto i al centro j, es decir el punto asociado a ese centro
    
    paraDividir=zeros(num_centros)
    nuevosCentros=zeros((R_n,num_centros))
    
    i=0   
    for i in range(num_puntos):
        fil=distancias[i,:]
        ind=argmin(fil)
        
        j=0
        for j in range(R_n):        
            nuevosCentros[(j,ind)]+=puntos[(j,i)]
            
        paraDividir[ind]+=1
    
    #Ahora los promedio diviendo en el numero de puntos que sumo para ese centro
    i=0
    for i in range(num_centros):
        div=paraDividir[i]
        
        if(div!=0):
            
            j=0
            for j in range(R_n):
                nuevosCentros[(j,i)]=nuevosCentros[(j,i)]/div
        
        else:
            
            j=0
            for j in range(R_n):
                nuevosCentros[(j,i)]=centros[(j,i)]
   
    #Ya tengo mis nuevos centros :D  
        
    return nuevosCentros   


#nuevosCentros: funcion que itera nuevosCentros hasta que sean iguales iteraciones n y n+1 
def nuevosCentrosIterados(centros,num_centros, puntos, num_puntos, R_n):
    
    nc_anterior=nuevosCentros(centros,num_centros, puntos, num_puntos, R_n)
    nc_actual=nuevosCentros(nc_anterior,num_centros, puntos, num_puntos, R_n)
    
    ant_igual_actual=False
    
    while(ant_igual_actual==False):
        nc_anterior=nc_actual
        nc_actual=nuevosCentros(nc_anterior,num_centros, puntos, num_puntos, R_n)
        
        contador=0
        i=0
        for i in range(R_n):
            j=0
            for j in range(num_centros):
                indice=(i,j)             
                if(nc_anterior[indice]==nc_actual[indice]):
                    contador+=1
                    
        if(contador==R_n*num_centros):
            ant_igual_actual=True
            break
        
    return nc_actual

#centrosAleatorios: Función para calcular K centros aleatorios
def centrosAletorios(R_n,K):
    cA=ones((R_n,K))
    
    i=0
    for i in range(K):
        j=0
        for j in range(R_n):
            indice=(j,i)
            cA[indice]=(-random.random())*(10**(-2))
            
    return cA
```

```python
#Ahora a hallar los K-Clusters para K=2

#K=2

K=2

mCA= centrosAletorios(2,K) 
nCI=nuevosCentrosIterados(mCA,K,misCoeficientes,4,2)

print "Para K=",K, "los centros en R_2 son: \n", nCI

i=0
for i in range(num_columnas):
    scatter(misCoeficientes[(0,i)], misCoeficientes[(1,i)], color='blue')
    
i=0    
for i in range(K):     
   scatter(nCI[(0,i)], nCI[(1,i)], color='red')
```
